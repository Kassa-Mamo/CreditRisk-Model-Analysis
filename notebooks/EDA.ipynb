```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Plot distribution for 'Amount' and 'Value'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Amount'], kde=True, color='blue', bins=50)
plt.title('Distribution of Amount')

plt.subplot(1, 2, 2)
sns.histplot(data['Value'], kde=True, color='green', bins=50)
plt.title('Distribution of Value')

plt.tight_layout()
plt.show()

# Plot distribution for 'FraudResult'
plt.figure(figsize=(6, 4))
sns.countplot(x='FraudResult', data=data, palette='viridis')
plt.title('Fraud Result Distribution (0 - No Fraud, 1 - Fraud)')
plt.show()
```

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Get summary statistics for numerical columns
summary_stats = numerical_data.describe()
print("Summary Statistics:")
print(summary_stats)

# Correlation between numerical variables
correlation_matrix = numerical_data.corr()

# Plot heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Convert all columns to numeric, forcing errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Select only numerical columns again after conversion
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Get summary statistics for numerical columns
summary_stats = numerical_data.describe()
print("Summary Statistics:")
print(summary_stats)

# Correlation between numerical variables
correlation_matrix = numerical_data.corr()

# Plot heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Handling missing data by filling with the median or dropping rows/columns
numerical_data.fillna(numerical_data.median(), inplace=True)

# Or if you'd like to drop columns with many missing values
# numerical_data.dropna(axis=1, thresh=int(0.5 * len(numerical_data)), inplace=True)
```

```python
# Convert 'TransactionStartTime' to datetime
data['TransactionStartTime'] = pd.to_datetime(data['TransactionStartTime'])

# Extract additional datetime features
data['TransactionYear'] = data['TransactionStartTime'].dt.year
data['TransactionMonth'] = data['TransactionStartTime'].dt.month
data['TransactionDay'] = data['TransactionStartTime'].dt.day
data['TransactionHour'] = data['TransactionStartTime'].dt.hour
data['TransactionMinute'] = data['TransactionStartTime'].dt.minute

# Check the first few rows after feature extraction
print(data[['TransactionStartTime', 'TransactionYear', 'TransactionMonth', 'TransactionDay', 'TransactionHour', 'TransactionMinute']].head())
```

```python
# Check for negative values in the 'Amount' column
negative_amounts = data[data['Amount'] < 0]

# Display the negative amounts
print(f"Negative Amounts: {negative_amounts.shape[0]} rows")
print(negative_amounts.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to categorical columns
categorical_columns = ['CurrencyCode', 'CountryCode', 'ProviderId', 'ProductId', 'ProductCategory', 'ChannelId', 'PricingStrategy']
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Check the first few rows after encoding
print(data[categorical_columns].head())
```
